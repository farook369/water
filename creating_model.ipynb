**Water Quality and Potability**


---


This dataset contains water quality measurements and assessments related to potability, which is the suitability of water for human consumption. The dataset's primary objective is to provide insights into water quality parameters and assist in determining whether the water is potable or no





```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/project/water_potability (1).csv')
df
```

```python
df.isna().sum()
```

```python
df['ph']=df['ph'].fillna(df['ph'].mean())
df['Sulfate']=df['Sulfate'].fillna(df['Sulfate'].mean())
df['Trihalomethanes']=df['Trihalomethanes'].fillna(df['Trihalomethanes'].mean())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
potable_count = df["Potability"].value_counts()[0]
non_potable_count = df["Potability"].value_counts()[1]
x= [potable_count, non_potable_count]
nlables=["Potable", "Non-potable"]
plt.figure(figsize=(5, 5))
plt.pie(x,labels=nlables,autopct="%1.1f%%")
plt.title("Potability Distribution")
plt.legend()
plt.show()
```

```python
X=df.iloc[:,:-1]
X
```

```python
y=df.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
X_scaled=scaler.fit_transform(X)
X_scaled
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_scaled,y,test_size=0.3,random_state=1)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

**Checking accuracy in differnt models**

```python
 from sklearn.neighbors import KNeighborsClassifier
 kn=KNeighborsClassifier()
 from sklearn.svm import SVC
 sv=SVC()
 from sklearn.naive_bayes import GaussianNB
 nb=GaussianNB()
 from sklearn.tree import DecisionTreeClassifier
 dt=DecisionTreeClassifier()
 from sklearn.ensemble import AdaBoostClassifier,RandomForestClassifier,GradientBoostingClassifier
 ab=AdaBoostClassifier()
 rf=RandomForestClassifier()
 gd=GradientBoostingClassifier()
 from xgboost import XGBClassifier
 xb=XGBClassifier()
 from sklearn.metrics import classification_report
 models=[kn,sv,nb,dt,ab,rf,gd,xb]
 for model in models:
  print("*****",model,"******")
  model.fit(X_train,y_train)
  y_pred=model.predict(X_test)
  print(classification_report(y_pred,y_test))


```

**Over sampling**

```python
from imblearn.over_sampling import SMOTE
os=SMOTE(random_state=1)
X_os,y_os=os.fit_resample(X,y)
```

```python
y_os.value_counts()
```

```python
Xos_scaled=scaler.fit_transform(X_os)
Xos_scaled
```

```python
XOS_train,XOS_test,yos_train,yos_test=train_test_split(Xos_scaled,y_os,test_size=0.3,random_state=1)
```

```python
XOS_train.shape,yos_train.shape
```

```python
XOS_test.shape,yos_test.shape
```

```python
for model in models:
  print("*******",model,"*******")
  model.fit(XOS_train,yos_train)
  y_pred1=model.predict(XOS_test)
  print(classification_report(yos_test,y_pred1))
```

**Under Sampling**

```python
from imblearn.under_sampling import RandomUnderSampler
us=RandomUnderSampler(random_state=1)
X_us,y_us=us.fit_resample(X,y)
```

```python
XUS_scaled=scaler.fit_transform(X_us)
XUS_scaled
```

```python
XUS_train,XUS_test,yus_train,yus_test=train_test_split(XUS_scaled,y_us,test_size=0.3,random_state=1)
```

```python
XUS_train.shape,yus_train.shape
```

```python
XUS_test.shape,yus_test.shape
```

```python
for model in models:
  print("*******",model,"*******")
  model.fit(XUS_train,yus_train)
  y_pred1=model.predict(XUS_test)
  print(classification_report(yus_test,y_pred1))
```
